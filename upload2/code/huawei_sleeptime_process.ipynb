```python
import os
import json
import re
from docx import Document
import pandas as pd
```

```python
# 数据路径
path = "../data/第2晚已监测完"
```

```python
#华为手表使用前后睡眠时长计算处理
dir = os.listdir(path)
name_list = []
before_huawei_sleep_time = []
after_huawei_sleep_time = []

for i in dir:
    dirpath = os.path.join(path,i)
    name  =re.findall(r'[（](.*?)[）]',str(i))
    
    hauwei_before =  os.path.join(dirpath,"Before_Use/Huawei_Watch_Data/Sleep_Report.json")
    with open(hauwei_before,'r')as f:
        a = json.load(f)
        t = a["healthRecords"][0]["value"][6]['integerValue']
        before_huawei_sleep_time.append(t)
        name_list.append(name[0])

    hauwei_after =  os.path.join(dirpath,"After_Use/Huawei_Watch_Data/Sleep_Report.json")
    with open(hauwei_after,'r')as f:
        a = json.load(f)
        t = a["healthRecords"][0]["value"][6]['integerValue']
        after_huawei_sleep_time.append(t)

df = pd.DataFrame()
df["姓名"] = name_list 
df["华为手表使用前"] = before_huawei_sleep_time
df["华为手表使用后"] = after_huawei_sleep_time

df.to_excel("../result/华为手表睡眠时长.xlsx")
```
