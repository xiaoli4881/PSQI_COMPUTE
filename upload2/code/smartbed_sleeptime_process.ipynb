```python
import os
import pandas as pd
import re
from docx import Document
```

```python
path = "../data/第2晚已监测完"
```

```python
dir = os.listdir(path)
list_name = []
list_before =[]
list_after = [] 
```

```python
#计算智能床使用前后睡眠时长
for i in dir:
    name = str(i)
    name = re.findall(r'[（](.*?)[）]',name)
    smart_bed_beofre = os.path.join(path,i,"Before_Use/Smart_Bed_Data/Sleep_Sign_Report_Data/Sleep_Report.csv")
    smart_bed_after = os.path.join(path,i,"After_Use/Smart_Bed_Data/Sleep_Sign_Report_Data/Sleep_Report.csv")
    df_before = pd.read_csv(smart_bed_beofre)

    df_after = pd.read_csv(smart_bed_after)
    before_bedtime = df_before["col_deep_sleep_duration"][0]+df_before["col_shallow_sleep_duration"][0]

    after_bedtime = df_after["col_deep_sleep_duration"][0]+df_after["col_shallow_sleep_duration"][0]

    list_name.append(name[0])
    list_before.append(before_bedtime)
    list_after.append(after_bedtime)
```

```python
dir = os.listdir(path)
list_light_before= []
list_light_after = []
before_duration = []
after_duration = []
```

```python
for i in dir:
    pathdir = os.path.join(path,i)
    name = re.findall(r'[（].*?[）]',str(i))
    name = name[0]
    #psg使用前睡眠时长
    before = os.path.join(pathdir,"Before_Use/Psg_Data/Hsy_Sleep_Report/Hsy_Sleep_Report1.docx")
    after = os.path.join(pathdir,"After_Use/Psg_Data/Hsy_Sleep_Report/Hsy_Sleep_Report1.docx")
    document = Document(before)  # 读入文件
    light_time = document.tables[1].rows[1].cells[0].text
    ligtht = light_time[-6:-1]
    list_light_before.append(ligtht)

    #psg使用后睡眠时长
    document = Document(after)
    light_time = document.tables[1].rows[1].cells[0].text
    ligtht = light_time[-6:-1]
    list_light_after.append(ligtht)
```

```python
# 保存成excel格式
df = pd.DataFrame()

df["姓名"] =list_name
df["psg使用前睡眠时长"] = list_light_before
df['智能床使用前睡眠时长'] = list_before
df ["psg使用后睡眠时长"] =list_light_after
df['智能床使用后睡眠时长'] = list_after
df.to_excel("../result/PSG智能床睡眠时长.xlsx")
```
